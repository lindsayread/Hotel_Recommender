```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    '''
    
    '''
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
def lsa_function(num_components, doc_text_matrix, vectorizer):
    '''
    Function that creates a LSA model based on a specified number of components, the document text matrix, and vectorizer.
    Prints the explained variance for each topic.
    Prints the top 15 words for each topic.
    Returns a topic-word matrix.
    '''
    lsa = TruncatedSVD(num_components)
    doc_topic = lsa.fit_transform(doc_text_matrix)
    print('Explained Variance:',lsa.explained_variance_ratio_)
    
    index = []
    for i in range(num_components):
        index.append(i)
    topic_word = pd.DataFrame(lsa.components_.round(3),
             index = index,
             columns = vectorizer.get_feature_names())
    
    print(display_topics(lsa, vectorizer.get_feature_names(), 15))
    
    return topic_word
```

```python
def nmf_function(num_components, doc_text_matrix, vectorizer):
    '''
    Function that creates a NMF model based on a specified number of components, the document text matrix, and vectorizer.
    Prints the top 15 words for each topic.
    Returns a topic-word matrix.
    '''
    nmf = NMF(num_components)
    doc_topic = nmf.fit_transform(doc_text_matrix)
    
    index = []
    for i in range(num_components):
        index.append(i)
    topic_word = pd.DataFrame(nmf.components_.round(3),
             index = index,
             columns = vectorizer.get_feature_names())
    
    print(display_topics(nmf, vectorizer.get_feature_names(), 15))
    
    return topic_word
```

```python
def nmf_HMatrix(num_components, doc_text_matrix, vectorizer):
    '''
    Function that creates a NMF model based on a specified number of components, document-term matrix, and vectorizer.
    Returns a document-topic matrix to show how much each document is represented by each topic.
    '''
    nmf = NMF(num_components)
    doc_topic = nmf.fit_transform(doc_text_matrix)
    
    idx = []
    for i in range(num_components):
        idx.append(i) 
    H = pd.DataFrame(doc_topic.round(3),
                    index = hotel_id,
                    columns = idx)
    return H
```
